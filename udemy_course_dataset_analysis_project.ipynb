```python
df.drop(columns='index',inplace=True)
```

```python
obj1=[]
num1=[]
for i in df:
  if df[i].dtype == 'int' or df[i].dtypes == 'float':
    num1.append(i)
  else:
    obj1.append(i)
```

```python
obj1,num1
```

```python
for i in num1:
  print(f"Mean of {i} = ",df[i].mean())
  print(f"Median of {i} = ",df[i].median())
  print(f"Mode of {i} = ",df[i].mode())
  print(f"Maximum of {i} = ",df[i].max())
  print(f"Minimum of {i} = ",df[i].min())
  print(f"Standard diviation  of {i} = ",df[i].std())
  print()
  print("*********************************************")
  print()
```

```python
for i in num1:
  plt.figure(figsize=(9,8))
  plt.boxplot(df[i])
  plt.title(f"Boxplot of {i}")
  plt.show()
```

```python
excep = ['course_title', 'url']

for i in obj1:
    for j in num1:
        if i not in excep and j not in excep:
            print(f"{i} and {j} Analysis")
            print("******************************")
            print(df.groupby(i)[j].agg(['mean', 'median', 'min', 'max', 'std', 'count']))
            print("___________________________________________________________________________")
```
